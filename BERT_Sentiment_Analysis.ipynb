```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")

data = ["I love you", "I hate you"]

sentiment_pipeline(data)
```

```python
data = ["The food is good", "The food is not good"]

sentiment_pipeline(data)
```
